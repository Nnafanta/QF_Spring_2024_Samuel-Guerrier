```python
%pip install pandas yfinance datetime numpy matplotlib.pyplot

import pandas as pd
import yfinance as yf
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt

# list of assets (updated with our chosen assets):
# Companies: Salesforce, HP, Lenovo, Ford, Tesla, Walmart, Target 
my_tickers = ['crm', 'hpq', 'lnvgy', 'f', 'tsla', 'wmt', 'tgt']

my_tickers = [ticker.upper() for ticker in my_tickers]


# Assign a start date and end date for the data download, 2 years of data
start_date = dt.datetime.today() - pd.Timedelta(740, "d")
end_date = dt.datetime.today()
```

```python
## Now download the data using the above variables, my_port is short for my portfolio
my_port = yf.download(start=start_date, end=end_date, tickers=my_tickers)
```

```python
## And review the data:
my_port.head()
```

```python
## We are only using Adj Close data
my_port = my_port['Adj Close']

my_port.head()
```

```python
## Calculate daily return of all stocks/assets.
d_returns = my_port.pct_change()
```

```python
## Define a function that takes in adjusted close price data or return data as inputs to calculate the mean, variance, and correlation matrix.
def find_mean_var_and_corr_matrix(input):   # input is either adjusted close price data or return data.
    return input.mean(), input.var(), input.corr()

    
    
```

```python
## Calculate the mean, variance, and correlation matrix by using adjusted close price data
find_mean_var_and_corr_matrix(my_port)
```

```python
## Calculate the mean, variance, and correlation matrix by using daily return data
find_mean_var_and_corr_matrix(d_returns)
```
