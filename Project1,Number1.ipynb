```python
import pandas as pd
import yfinance as yf
import datetime as dt

# list of assets (updated with our chosen assets):
# Companies: Salesforce, HP, Lenovo, Ford, Tesla, Walmart, Target 
my_tickers = ['CRM', 'HPQ', 'LNVGY', 'F', 'TSLA', 'WMT', 'TGT']


# Assign a start date and end date for the data download, 1 year of data
start_date = dt.datetime.today() - pd.Timedelta(400, "d")
end_date = dt.datetime.today()
```

```python
## Now download the data using the above variables, my_port is short for my portfolio
my_port = yf.download(start=start_date, end=end_date, tickers=my_tickers)
```

```python
## And review the data:
my_port.head()
```

```python
## We may only need one index of the multi-index that was returned (Adj Close)
my_port = my_port['Adj Close']

my_port.head()
```

```python

my_port_table = pd.DataFrame(index = my_tickers)
    
```

```python
## Calculation for 1a. and 1b. in the 1st project. Column of the tickers and portfolio weights.

port_weight = [1/len(my_tickers)] * len(my_tickers)
my_port_table['Portfolio_Weight'] = port_weight

my_port_table

```

```python
## Calculation for 1c. 
def get_stock_prices(my_tickers, start_date, end_date):
    my_port
    return my_port

stock_prices = get_stock_prices(my_tickers, start_date, end_date)
returns_3months = stock_prices.pct_change().rolling(window=63).sum().dropna()
volatility = returns_3months.std() * (252 ** 0.5)
result_df = pd.DataFrame({'Ticker': my_tickers, 'Volatility': volatility})

print(result_df)
```

```python
my_port_table['Annualized Volatility'] = volatility
my_port_table
```

```python
## Calculation for 1d.

## First download Adjusted Close Price data for S&P 500 ETF
spy = yf.download(start=start_date, end=end_date, tickers='SPY')
spy = spy['Adj Close']

## Combine portfolio and SPY data into a single DataFrame
my_port_and_spy_data = pd.concat([my_port, spy], axis=1)
my_port_and_spy_data.columns = my_tickers + ['SPY']

## Calculate daily returns
my_port_and_spy__d_returns = my_port_and_spy_data.pct_change().dropna()

## Calculate covariance matrix
my_port_and_spy_cov_matrix = my_port_and_spy__d_returns.cov()

## Extract covariances with SPY
cov_with_spy = my_port_and_spy_cov_matrix['SPY'].drop('SPY')

## Calculate SPY variance
spy_variance = my_port_and_spy_cov_matrix.loc['SPY', 'SPY']

# Calculate beta
betas_spy = cov_with_spy / spy_variance

spy_beta_table = pd.DataFrame({'Tickers': my_tickers, 'Beta': betas_spy})
print(spy_beta_table)


'''
## Then calculate the daily returns of portfolio and S&P 500 ETF
my_port_d_returns = prices.pct_change().dropna()
spy_d_returns = spy.pct_change().dropna()

covariances = my_port_d_returns.covwith(spy_d_returns)
'''

```

```python
## Resulting table after calculating 1d. 
my_port_table['Beta against SPY'] = betas_spy
my_port_table
```

```python
## Calculation for 1e.

## First download Adjusted Close Price data for S&P 500 ETF
iwm = yf.download(start=start_date, end=end_date, tickers='IWM')
iwm = iwm['Adj Close']

## Combine portfolio and IWM data into a single DataFrame
my_port_and_iwm_data = pd.concat([my_port, iwm], axis=1)
my_port_and_iwm_data.columns = my_tickers + ['IWM']

## Calculate daily returns
my_port_and_iwm__d_returns = my_port_and_iwm_data.pct_change().dropna()

## Calculate covariance matrix
my_port_and_iwm_cov_matrix = my_port_and_iwm__d_returns.cov()

## Extract covariances with IWM
cov_with_iwm = my_port_and_iwm_cov_matrix['IWM'].drop('IWM')

## Calculate IWM variance
iwm_variance = my_port_and_iwm_cov_matrix.loc['IWM', 'IWM']

# Calculate beta
betas_iwm = cov_with_iwm / iwm_variance







```

```python
## Resulting table after calculating 1e.
my_port_table['Beta against IWM'] = betas_iwm
my_port_table
```

```python
## Calculation for 1f.

## First download Adjusted Close Price data for DIA
dia = yf.download(start=start_date, end=end_date, tickers='DIA')
dia = dia['Adj Close']

## Combine portfolio and DIA data into a single DataFrame
my_port_and_dia_data = pd.concat([my_port, dia], axis=1)
my_port_and_dia_data.columns = my_tickers + ['DIA']

## Calculate daily returns
my_port_and_dia__d_returns = my_port_and_dia_data.pct_change().dropna()

## Calculate covariance matrix
my_port_and_dia_cov_matrix = my_port_and_dia__d_returns.cov()

## Extract covariances with IWM
cov_with_dia = my_port_and_dia_cov_matrix['DIA'].drop('DIA')

## Calculate DIA variance
dia_variance = my_port_and_dia_cov_matrix.loc['DIA', 'DIA']

# Calculate beta
betas_dia = cov_with_dia / dia_variance
```

```python
## Resulting table after calculating 1f.
my_port_table['Beta against DIA'] = betas_dia
my_port_table
```

```python
## Calculation of 1g., Average Weekly Drawdown, (52-week Low minus 52-week High) / 52-week High

av_w_drawdown = (my_port.rolling(5).min() - my_port.rolling(5).max()).mean() 

my_port_table['Average Weekly Drawdown'] = av_w_drawdown 
my_port_table
```

```python
## Calculation of 1h., Maximum Weekly Drawdown (52-week Low minus 52-week High) / 52-week High
max_w_drawdown = (my_port.rolling(5).min() - my_port.rolling(5).max()).max()

my_port_table['Maximum Weekly Drawdown'] = max_w_drawdown 
my_port_table
```

```python
## Calculation of 1i., Total Return (using trailing 10-years)

start_date = dt.datetime.today() - pd.Timedelta(3650, "d")  # First, put 3650, which represents 10 years from today.
my_port = yf.download(start=start_date, end=end_date, tickers=my_tickers)  # Then download data for 10 years
my_port = my_port['Adj Close']
t_return = my_port.pct_change().sum()[-20:] # Calculate the yearly returns for each stock, then add up the yearly returns of each stock, then only take into account the last 20 days of data for a trailing calculation
t_return

```

```python
## Table with 1i included.
my_port_table['Total Return'] = t_return
my_port_table
```

```python
## Calculation of 1j., Annualized Total Return (using trailing 10-years)

t_return_ann = ((my_port.pct_change().sum()[-20:] + 1) ** 252) - 1 
t_return_ann
```

```python
## Stock porfolio table with 1j included.
my_port_table['Annualized Total Return'] = t_return_ann
my_port_table
```
